#### Question Text - Course Information

Write a program that creates a dictionary containing course numbers and the room numbers of the rooms where the courses meet. The dictionary should have the following keyvalue pairs:



The program should also create a dictionary containing course numbers and the names of the instructors that teach each course. The dictionary should have the following key-value pairs:



The program should also create a dictionary containing course numbers and the meeting times of each course. The dictionary should have the following key-value pairs:



The program should let the user enter a course number, then it should display the course’s room number, instructor, and meeting time.
#### High-Level Overview

First I created three dictionaries for the room number, instructor and meeting times that correspond to each course. Next I prompted the user for a course number to display details. Lastly I displayed the room number, instructor and meeting time to the user.

```python
# Create dictionaries for user input 
roomNumber = {'CS101':3004, 'CS102':4501, 'CS103':6755, 'NT110':1244, 'CM241':1411}
instructor = {'CS101':'Haynes', 'CS102':'Alvarado', 'CS103':'Rich', 'NT110':'Burke', 'CM241':'Lee'}
time = {'CS101':'8:00 a.m.', 'CS102':'9:00 a.m.', 'CS103':'10:00 a.m.', 'NT110':'11:00 a.m', 'CM241':'1:00 p.m.'}
# prompt user for course number to display details
userInput = input("Enter a course number to display details: ")
#display details to user
print(f"The room number is: {roomNumber[userInput]}")
print(f"The instructor  is: {instructor[userInput]}")
print(f"The meeting time is: {time[userInput]}")
```
