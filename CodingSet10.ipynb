### Problem **1**

#### Question Text - Pet Class




#### High-Level Overview

First I created the pet class and initialized the attributes. Next, I made the accessor and mutator methods for each attribute. Afterwards, I defined the main function and called the pet class object after prompting the user for the pet details. Lastly, displayed the results and called the main function.

```python
class Pet:
    # initialize attributes
    def __init__(self, name, animal_type, age):
        self.__name = name
        self.__animal_type = animal_type
        self.__age = age
    # set_name method
    def set_name(self, name):
        self.__name = name
    # set_animal_type method
    def set_animal_type(self, animal_type):
        self.__animal_type = animal_type
    # set_age method
    def set_age(self, age):
        self.__age = age
    # get_name method
    def get_name(self):
        return self.__name
    # get_animal_type method
    def get_animal_type(self):
        return self.__animal_type
    # get_age method
    def get_age(self):
        return self.__age
    def __str__(self):
        return f'Name: {self.__name}\nAnimal Type: {self.__animal_type}\nAge: {self.__age}'
# define main function
def main():
    # prompt user
    name = input("Enter the name of your pet: ")
    animal_type =  input("What type of animal is your pet? ")
    age = input("How old is your pet?")
    # assign attributes to object thats created
    myPet = Pet(name, animal_type, age)
    # display results
    print(f"Your pet\'s name is: {myPet.get_name()}\nYour pet is a {myPet.get_animal_type()}\nYour pet is {myPet.get_age()} years old.\n")
# call main function
if __name__ == "__main__":
    main()
```

### Problem **2**

#### Question Text - Car Class




#### High-Level Overview

First I created the car class and initialized it's attributes. Next, I defined the accelerate, brake, and accessor method for speed. I then made the main function, prompted the user for car details and made the car object. Next, I called the accelerate and brake functions 5 times respectively and displayed the speed at each interval using two for loops. Lastly I called the main function.


```python
# create car class
class Car:
    # define/initialize attributes
    def __init__(self, year_model, make):
        self.__year_model = year_model
        self.__make = make
        self.__speed = 0
    # define accelerate
    def accelerate(self):
        self.__speed += 5
    # define brake
    def brake(self):
        # decrement speed
        self.__speed -= 5
        # check if speed is less than 0
        if self.__speed < 0:
            self.__speed = 0
    # define accessor for speed
    def get_speed(self):
        return self.__speed
# define main function
def main():
    # create car object after prompting user
    make = input("Enter the make of the car: ")
    model = input("Enter the model of the care: ")
    car1 = Car(model, make)
    # acclerate 5 times and display speed at each acceleration
    print("accelerating...")
    for i in range(5):
        car1.accelerate()
        print(f'Current speed: {car1.get_speed()} mph')
    # brake 5 times and display speed at each deceleration
    print("slowing down...")
    for i in range(5):
        car1.brake()
        print(f'Current speed: {car1.get_speed()} mph')
if __name__ == "__main__":
    main()
```
